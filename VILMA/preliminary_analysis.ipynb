```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet('yellow_tripdata_2024-01.parquet')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
print(df.shape[0])
df = df.dropna(axis=0)
print(df.shape[0])
df = df[df.passenger_count > 0]
print(df.shape[0])
df = df[df.trip_distance > 0]
print(df.shape[0])
df = df[df.fare_amount > 0]
print(df.shape[0])
df = df[df.total_amount > 0]
print(df.shape[0])
```

```python
df['year'] = df['tpep_pickup_datetime'].dt.year
df['month'] = df['tpep_pickup_datetime'].dt.month
df['day'] = df['tpep_pickup_datetime'].dt.day
df['hour'] = df['tpep_pickup_datetime'].dt.hour
df['day_of_week'] = df['tpep_pickup_datetime'].dt.dayofweek

df
```

```python
df['trip_duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df['trip_duration'] = df.trip_duration.dt.total_seconds()
print(df.shape[0])
df = df[df.trip_duration > 0]
print(df.shape[0])
df
```

```python
# Trips lasting longer than 24 hours
df[df.trip_duration > 60*60*24]
```

```python
plt.figure(figsize=(16,8))
for i, col in enumerate(['VendorID', 'passenger_count', 'trip_distance', 'RatecodeID',        'PULocationID', 'DOLocationID', 'payment_type', 'fare_amount', 'extra',
       'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge',
       'total_amount', 'congestion_surcharge', 'Airport_fee']):
    
    plt.subplot(4,4,i+1)
    plt.hist(df[col])
    plt.title(f'{col}')
plt.tight_layout()
```
