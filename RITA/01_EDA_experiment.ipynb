```python
# List for all the parquet files 
file_paths = [f.path for f in parquets if f.path.endswith(".parquet")]
```

```python
file_paths
```

```python
# Function to extract the month from the filename
def extract_year_month(filename):
    filename= filename.replace('.parquet','') # To remove .parquet

    # Get the file name from the full path
    file_name = filename.split('/')[-1]  # ex: 'yellow_tripdata_2024_01'

    # Then split the name
    name_split= file_name.split('_')

    # Get the year and month
    year= name_split[2]
    month= name_split[3]
    
    return year,month
```

```python
# Loads the files and adds the 'month' column
dfs = []
for path in file_paths:
    year, month = extract_year_month(path)
    df = spark.read.parquet(path)
    df = df.withColumn("month", F.lit(int(month)))  # adds the column month
    df= df.withColumn("year", F.lit(int(year))) 
    dfs.append(df)
```

```python
# Joins everything into a single DataFrame
full_df = reduce(DataFrame.unionByName, dfs)
```

```python
cat_cols = [col for col in categorical if sample_df[col].nunique() <= 20]
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2_corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    r_corr = r - ((r-1)**2)/(n-1)
    k_corr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))
```

```python
cramer_matrix = pd.DataFrame(index=cat_cols, columns=cat_cols)

for col1 in cat_cols:
    for col2 in cat_cols:
        if col1 == col2:
            cramer_matrix.loc[col1, col2] = 1.0
        else:
            try:
                cramer_matrix.loc[col1, col2] = cramers_v(sample_df[col1], sample_df[col2])
            except:
                cramer_matrix.loc[col1, col2] = np.nan

cramer_matrix = cramer_matrix.astype(float)

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cramer_matrix, annot=True, cmap="YlGnBu", fmt=".2f", square=True)
plt.title("Cramrs V Association Between Categorical Variables")
plt.tight_layout()
plt.show()
```

###  Interpretation: Cramrs V Association Between Categorical Variables

Cramrs V is designed to detect association patterns between categorical variables, and low values here indicate no strong relationship between them.

The heatmap shows that the categorical variables have **no significant association** with each other  as expected, the diagonal contains `1.00` values (each variable is perfectly associated with itself), while all off-diagonal values are near zero.

This result suggests that variables are **statistically independent** in this dataset. The values of one variable do **not influence or depend** on the values of another.

## Numerical vs Categorical

```python
def eta_squared(df, num_col, cat_col):
    # Keep only the relevant columns and drop missing values
    df = df[[num_col, cat_col]].dropna()

    # Convert the categorical column to a proper category type
    categories = df[cat_col].astype("category")

    # Extract the numerical column (the dependent variable)
    y = df[num_col]

    # Create a list of numerical values grouped by category
    groups = [y[categories == cat] for cat in categories.cat.categories]

    # Compute the Between-Group Sum of Squares (SS_between)
    ss_between = sum(len(g) * (g.mean() - y.mean())**2 for g in groups)

    # Compute the Total Sum of Squares (SS_total)
    ss_total = sum((y - y.mean())**2)

    # Return Eta Squared: the proportion of variance explained by the categorical variable
    return ss_between / ss_total if ss_total != 0 else np.nan
```

```python
results = []

for cat in categorical:
    for num in numeric_cols:
        try:
            score = eta_squared(sample_df.dropna(subset=[cat, num]), num, cat)
            results.append((cat, num, round(score, 3)))
        except:
            results.append((cat, num, None))

eta_df = pd.DataFrame(results, columns=["Categorical", "Numerical", "Eta"])
display(eta_df)
```

```python
eta_df["Eta"] = pd.to_numeric(eta_df["Eta"], errors="coerce")

eta_pivot = eta_df.pivot(index="Categorical", columns="Numerical", values="Eta")

print(eta_pivot)

plt.figure(figsize=(10, 6))
sns.heatmap(eta_pivot, annot=True, cmap="YlGnBu", fmt=".2f", linewidths=0.5, mask=eta_pivot.isnull())
plt.title("Eta Squared ()  Association Between Categorical and Numerical Variables")
plt.tight_layout()
plt.show()
```

###  Interpretation of Categorical vs Numerical Association (Eta Squared)

The heatmap above visualizes the strength of association between categorical and numerical variables using Eta Squared. This metric quantifies the proportion of variance in a numerical variable that is explained by a categorical variable, ranging from 0 (no association) to 1 (perfect association).

```python
from scipy.stats import kruskal
import pandas as pd
import numpy as np

# Lista onde vamos guardar os resultados
kruskal_results = []

# Loop pelos pares
for cat in categorical:
    for num in numeric_cols:
        try:
            df_temp = sample_df[[cat, num]].dropna()
            groups = df_temp.groupby(cat)[num].apply(list)
            if len(groups) >= 2:
                stat, p = kruskal(*groups)
                kruskal_results.append((cat, num, round(p, 4)))
            else:
                kruskal_results.append((cat, num, None))
        except:
            kruskal_results.append((cat, num, None))

# DataFrame de resultados
kruskal_df = pd.DataFrame(kruskal_results, columns=["Categorical", "Numerical", "Kruskal_p"])

# Exibir
display(kruskal_df)
```

```python
# Converte p-valores para numrico (garante compatibilidade)
kruskal_df["Kruskal_p"] = pd.to_numeric(kruskal_df["Kruskal_p"], errors="coerce")

# Pivot para heatmap
kruskal_pivot = kruskal_df.pivot(index="Categorical", columns="Numerical", values="Kruskal_p")
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(kruskal_pivot, annot=True, cmap="Reds_r", fmt=".3f", linewidths=0.5, mask=kruskal_pivot.isnull(), cbar_kws={'label': 'p-value'})
plt.title("Kruskal-Wallis Test  p-values for Categorical  Numerical")
plt.tight_layout()
plt.show()
```

###  Kruskal-Wallis Test  Interpretation of Categorical vs Numerical Relationship

The heatmap above shows the **p-values** resulting from the Kruskal-Wallis test, used to assess whether there are statistically significant differences in the distributions of numerical variables across the groups defined by each categorical variable.

####  Key Observations:

- **Almost all combinations show extremely low p-values ( 0.000)**, indicating **strong statistical evidence** that the distributions of numerical variables **differ significantly across categories**.
- The only exception is:
  - `store_and_fwd_flag` vs `congestion_surcharge` (p = 0.158)
    This value exceeds the typical 0.05 threshold, suggesting **no significant difference** in congestion surcharge between the 'store and forward' groups.
- Variables such as `PULocationID`, `RatecodeID`, and `payment_type` consistently show p-values near 0, confirming their **significant influence on fare-related numerical features**.

####  Conclusion:

The Kruskal-Wallis test confirms that **most categorical variables significantly affect the distribution of fare-related numerical fields**. These results align with the findings from the Eta Squared analysis, reinforcing that location and rate-based features are important drivers in the dataset. Only the `store_and_fwd_flag` appears to lack statistical relevance for most numerical outcomes.
